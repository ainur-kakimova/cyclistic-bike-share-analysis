```r
CREATE OR REPLACE TABLE `cosmic-answer-443416-r8.Cyclistic.Biking_cleaned` AS
SELECT *
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_all_months`
WHERE 
  ride_id IS NOT NULL
  AND started_at IS NOT NULL
  AND ended_at IS NOT NULL
  AND member_casual IS NOT NULL
  AND start_station_name IS NOT NULL
  AND end_station_name IS NOT NULL
  AND start_lat IS NOT NULL AND start_lng IS NOT NULL
  AND end_lat IS NOT NULL AND end_lng IS NOT NULL
  AND start_lat != 0 AND start_lng != 0
  AND end_lat != 0 AND end_lng != 0
  AND TIMESTAMP_DIFF(ended_at, started_at, SECOND) > 0
  AND TIMESTAMP_DIFF(ended_at, started_at, HOUR) < 24;
```

The above query removes nulls and extreme outliers.

```r
CREATE OR REPLACE TABLE `cosmic-answer-443416-r8.Cyclistic.Biking_with_duration` AS
SELECT 
  *,
  TIMESTAMP_DIFF(ended_at, started_at, MINUTE) AS ride_duration_minutes
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_cleaned`;
```

The above query calculates the ride duration in minutes

```r
SELECT *
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_with_duration`
ORDER BY started_at ASC;
```

The above query organizes data in ascending order based on the column "started_at"

```r
CREATE OR REPLACE TABLE `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated` AS
SELECT *
FROM (
  SELECT *,
         ROW_NUMBER() OVER (PARTITION BY ride_id ORDER BY started_at ASC) AS row_num
  FROM `cosmic-answer-443416-r8.Cyclistic.Biking_with_duration`
)
WHERE row_num = 1;
```

The above query is used for deduplication

```r
SELECT ride_id, COUNT(*) AS count
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated`
GROUP BY ride_id
HAVING count > 1;
```

The above query checks for duplicates.

```r
SELECT 
  member_casual,
  ROUND(AVG(ride_duration_minutes), 2) AS avg_duration
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated`
GROUP BY member_casual;
```

This code calculates the average ride duration in minutes for each user type (member or casual) in the Cyclistic dataset.

```r
SELECT 
  member_casual,
  FORMAT_DATE('%A', DATE(started_at)) AS weekday,
  ROUND(AVG(ride_duration_minutes), 2) AS avg_duration,
  COUNT(*) AS total_rides
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated`
GROUP BY member_casual, weekday
ORDER BY weekday;
```

This code calculates the average ride duration and total number of rides for each user type by weekday in the Cyclistic dataset.

```r
SELECT 
  member_casual,
  rideable_type,
  COUNT(*) AS ride_count
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated`
GROUP BY member_casual, rideable_type;
```

This code counts the number of rides for each combination of user type and bike type in the Cyclistic dataset.

```r
CREATE OR REPLACE TABLE `cosmic-answer-443416-r8.Cyclistic.Frequency_by_user_day_hour` AS
SELECT 
  member_casual,
  FORMAT_DATE('%A', DATE(started_at)) AS weekday_name,
  EXTRACT(HOUR FROM started_at) AS hour_of_day,
  COUNT(*) AS ride_frequency
FROM `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated`
GROUP BY member_casual, weekday_name, hour_of_day;
```

This code creates a new table that shows the ride frequency for each user type by weekday and hour of the day in the Cyclistic dataset.

```r
CREATE OR REPLACE TABLE `cosmic-answer-443416-r8.Cyclistic.Biking_for_tableau` AS
SELECT 
  d.ride_id,
  d.rideable_type,
  d.started_at,
  d.ended_at,
  d.start_station_name,
  d.start_station_id,
  d.end_station_name,
  d.end_station_id,
  d.start_lat,
  d.start_lng,
  d.end_lat,
  d.end_lng,
  d.member_casual,
  
  -- Calculated fields
  TIMESTAMP_DIFF(d.ended_at, d.started_at, MINUTE) AS ride_duration_minutes,
  EXTRACT(MONTH FROM d.started_at) AS month_number,
  FORMAT_DATE('%B', DATE(d.started_at)) AS month_name,
  EXTRACT(DAYOFWEEK FROM d.started_at) AS weekday_number,
  FORMAT_DATE('%A', DATE(d.started_at)) AS weekday_name,
  EXTRACT(HOUR FROM d.started_at) AS hour_of_day,

  -- Joined frequency
  f.ride_frequency AS frequency_per_user_time

FROM `cosmic-answer-443416-r8.Cyclistic.Biking_deduplicated` d
LEFT JOIN `cosmic-answer-443416-r8.Cyclistic.Frequency_by_user_day_hour` f
  ON d.member_casual = f.member_casual
  AND FORMAT_DATE('%A', DATE(d.started_at)) = f.weekday_name
  AND EXTRACT(HOUR FROM d.started_at) = f.hour_of_day;
```

This code creates a Tableau-ready table by combining ride details with time-based features and ride frequency by user type, weekday, and hour.
